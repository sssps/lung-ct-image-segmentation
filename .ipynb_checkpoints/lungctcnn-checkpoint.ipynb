```python
import os
import random
from glob import glob
from keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
from keras import preprocessing
from keras.preprocessing import image as image_utils
%matplotlib inline
```

```python
class_names = ['NORMAL','CANCER']
```

```python
width = 96
height = 96
```

```python
import os
import random
from glob import glob
from keras import preprocessing 
```

```python
def load_images(base_path):
    images = []
    path = os.path.join(base_path, '*.jpeg')
    for image_path in glob(path):
        image = preprocessing.image.load_img(image_path,
                                             target_size=(width, height))
        x = preprocessing.image.img_to_array(image)

        images.append(x)
    return images
```

```python
images_type_1 = load_images('/content/drive/My Drive/Predicitng-Lungs-Disease--master/dataset/chest_xray/train/NORMAL')
images_type_2 = load_images('/content/drive/My Drive/Predicitng-Lungs-Disease--master/dataset/chest_xray/train/CANCER')
```

```python
plt.figure(figsize=(12,8))

for i in range(3):
    plt.subplot(1, 3, i+1)
    image = preprocessing.image.array_to_img(random.choice(images_type_1))
    plt.imshow(image)
    
    plt.axis('off')
    plt.title('{} image'.format(class_names[0]))

# show the plot
plt.show()
```

```python
plt.figure(figsize=(12,8))

for i in range(3):
    plt.subplot(1, 3, i+1)
    image = preprocessing.image.array_to_img(random.choice(images_type_2))
    plt.imshow(image)
    
    plt.axis('off')
    plt.title('{} image'.format(class_names[1]))

# show the plot
plt.show()
```

```python
X_type_1 = np.array(images_type_1)
X_type_2 = np.array(images_type_2)

print(X_type_1.shape)
print(X_type_2.shape)
```

```python
X = np.concatenate((X_type_1, X_type_2), axis=0)
```

```python
X = X / 255.

X.shape
```

```python
from keras.utils import to_categorical

y_type_1 = [0 for item in enumerate(X_type_1)]
y_type_2 = [0 for item in enumerate(X_type_2)]


y = np.concatenate((y_type_1, y_type_2), axis=0)

y = to_categorical(y, num_classes=len(class_names))

print(y.shape)
```

```python
from keras.models import Sequential
from keras.layers.core import Activation, Dropout, Flatten, Dense
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from keras.optimizers import Adam

# default parameters
conv_1 = 16
conv_1_drop = 0.2
conv_2 = 32
conv_2_drop = 0.2
dense_1_n = 1000
dense_1_drop = 0.2
dense_2_n = 550
dense_2_drop = 0.2
lr = 0.001

epochs = 30
batch_size = 32
color_channels = 3

def build_model(conv_1_drop=conv_1_drop, conv_2_drop=conv_2_drop,
                dense_1_n=dense_1_n, dense_1_drop=dense_1_drop,
                dense_2_n=dense_2_n, dense_2_drop=dense_2_drop,
                lr=lr):
    model = Sequential()

    model.add(Convolution2D(conv_1, (3, 3),
                            input_shape=(width, height, color_channels),
                            activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(conv_1_drop))

    model.add(Convolution2D(conv_2, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(conv_2_drop))
        
    model.add(Flatten())
        
    model.add(Dense(dense_1_n, activation='relu'))
    model.add(Dropout(dense_1_drop))

    model.add(Dense(dense_2_n, activation='relu'))
    model.add(Dropout(dense_2_drop))

    model.add(Dense(len(class_names), activation='softmax'))

    model.compile(loss='categorical_crossentropy',
                  optimizer=Adam(lr=lr),
                  metrics=['accuracy'])

    return model
```

```python
import numpy as np
np.random.seed(1) # for reproducibility

# model with base parameters
model = build_model()

model.summary()
```

```python
epochs = 10
```

```python
model.fit(X, y, epochs=epochs,verbose=1)
```

```python
model.save("lungmodel.h5")
print("Saved model to disk")
```
